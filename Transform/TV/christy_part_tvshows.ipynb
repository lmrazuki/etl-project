```python
# Dependencies and Setup
import pandas as pd

# Read Purchasing File and store into Pandas data frame
tv = pd.read_csv('../../Extract/TV/tv_shows.csv', encoding='latin1')

tv.head()
```

```python
tv = tv.loc[tv['Prime Video']==0]
tv.head()
```

```python
# drop non-used columns
tv = tv.drop(columns = ['Unnamed: 0', 'IMDb', 'Rotten Tomatoes', 'type', 'Prime Video'])
tv.head()
```

```python
# Rename Disney+ to Disney
tv = tv.rename(columns={'Disney+': 'Disney'})
tv.head()
```

```python
# Check/drop if there is duplicated movies
tv = tv.drop_duplicates('Title', keep='last').reset_index(drop=True)
tv.head()
```

```python
# check how many rows are left in the dataframe
tv['Year'].count()
```

```python
# combine the title and year into key
tv["Key"] = tv["Title"] + tv["Year"].astype(str)
tv
```

```python
# Export to csv file
tv.to_csv("../../Transform/TV/tv_cw.csv")
```

```python
# read the other csv data
tv2 = pd.read_csv("../../Transform/TV/TV_df_omdb.csv")
tv2.head()
```

```python
# Merging two csv files into one
merged = tv.merge(tv2, on='Key')
merged.head()
```

```python
# drop duplicated columns
merged = merged.drop(columns = ['Unnamed: 0', 'Title_y', 'Year_y'])
merged = merged.rename(columns={'Title_x': 'Title', 'Year_x': 'Year'})
merged.head()
```

```python
merged.to_csv("../../Load/TV/output_TV_combined.csv", index=False)
```
