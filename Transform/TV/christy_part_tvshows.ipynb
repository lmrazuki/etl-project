```python
# Dependencies and Setup
import pandas as pd

# Read Purchasing File and store into Pandas data frame
tv = pd.read_csv("Extract/TV/tv_shows.csv", encoding='latin1')

tv.head()
```

```python
tv = tv.loc[tv['Prime Video']==0]
tv.head()
```

```python
# drop non-used columns
tv = tv.drop(columns = ['Unnamed: 0', 'IMDb', 'Rotten Tomatoes', 'type', 'Prime Video'])
tv.head()
```

```python
# Rename Disney+ to Disney, rename Unnamed: 0 to ID
tv = tv.rename(columns={'Disney+': 'Disney', 'Unnamed: 0': 'ID'})
tv.head()
```

```python
# Check/drop if there is duplicated movies
tv = tv.drop_duplicates('Title', keep='last').reset_index(drop=True)
tv.head()
```
