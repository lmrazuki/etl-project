```python
# Dependencies and Setup
import pandas as pd

# Read File and store into Pandas data frame
movies = pd.read_csv("../../Extract/Movies/movies.csv", encoding='latin1')

movies.head()
```

```python
movies = movies.loc[movies['Prime Video']==0]
movies.head()
```

```python
# drop non-used columns
movies = movies.drop(columns = ['Unnamed: 0', 'ID', 'IMDb', 'Rotten Tomatoes', 'Type', 'Directors', 
                                'Genres', 'Country', 'Language', 'Runtime', 'Prime Video'])
movies.head()
```

```python
# Rename Disney+ to Disney
movies = movies.rename(columns={'Disney+': 'Disney'})
movies.head()
```

```python
# Check/drop if there is duplicated movies
movies = movies.drop_duplicates('Title', keep='last').reset_index(drop=True)
movies.head()
```

```python
# check how many rows are left in the dataframe
movies['Year'].count()
```

```python
# combine the title and year into key
movies["key"] = movies["Title"] + movies["Year"].astype(str)
movies.head()
```

```python
# Export to csv file
movies.to_csv("../../Transform/Movies/movies_cw.csv")
```

```python
# read the other csv data
movies2 = pd.read_csv('../../Transform/Movies/movies_df_clean.csv')
movies2.head()
```

```python
# Merging two csv files into one
merged = movies.merge(movies2, on='key')
merged.head()
```

```python
# drop duplicated columns
merged = merged.drop(columns = ['Unnamed: 0', 'Title_y', 'Year_y'])
merged = merged.rename(columns={'Title_x': 'Title', 'Year_x': 'Year'})
merged.head()
```

```python
merged.to_csv("../../Load/Movies/output_movie_combined.csv", index=False)
```
