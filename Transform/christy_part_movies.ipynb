```python
# Dependencies and Setup
import pandas as pd

# Read Purchasing File and store into Pandas data frame
movies = pd.read_csv("Extract/Movies/movies.csv", encoding='latin1')

movies.head()
```

```python
movies.count()
```

```python
movies = movies.loc[movies['Prime Video']==0]
movies.head()
```

```python
# drop non-used columns
movies = movies.drop(columns = ['Unnamed: 0', 'ID', 'IMDb', 'Rotten Tomatoes', 'Type', 'Directors', 
                                'Genres', 'Country', 'Language', 'Runtime', 'Prime Video'])
movies.head()
```

```python
# Rename Disney+ to Disney
movies = movies.rename(columns={'Disney+': 'Disney'})
movies.head()
```

```python
# Check/drop if there is duplicated movies
movies = movies.drop_duplicates('Title', keep='last').reset_index(drop=True)
movies.head()
```
